```python
independent = data.drop('class', axis=1)
independent.columns.value_counts().sum()

Input = independent.values
# Input = np.percentile(Input, float(U9))
Input = Input.astype('float')
Output = data['class'].values
Output = Output.astype('float')


type(Output)
```

```python
data.isnull().sum()
```

```python
from sklearn import feature_selection
from sklearn.linear_model import LinearRegression

feature_selection.f_regression(Input, Output, center=True)
model = LinearRegression(fit_intercept=True)

selector = feature_selection.RFE(model, n_features_to_select=10, step=1)
selector = selector.fit(Input, Output)
selector.support_
```

```python
selector.ranking_
```
